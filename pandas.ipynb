```python
# -7% betwee 2013 and 2020 CIA, select cia13 and cia20 country
def percent_difference(country):
  percent = countries.CIA13-countries.CIA20
  return percent > 7
countries.loc[lambda country: percent_difference(country), ["country","CIA13","CIA20"]]

```
