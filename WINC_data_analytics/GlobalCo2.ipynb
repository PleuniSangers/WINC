```python
#assignment global CO2 emission setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('max_rows', 99999) 
pd.set_option('max_colwidth', 400)
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions")
emission= tables[1]
emission.columns= ["country","1990","2005","2017","%world17","2017v1990","arealand17","capital17","2018lucf","2018"]
emission = emission.loc[3:]
emission = emission.drop(index=67) #european union isn't a country
emission
```

```python
#graph 1 CO2 of the bigger countries part I 5 biggest countries latest measurement. Not sure if you want with or without LUCF so for now I chose 2018 without LUCF for this excersize
#first step remove null 
emission = emission[~emission['2018'].isnull()]
#check datatypes, change object into float. for 169 changed it in Serbia en 45.45, because it can't have 2 values seperated / .
#emission.loc[169,["country","2018"]] = ["serbia","45.45"]
emission["2018"]=emission["2018"].astype(float)
emission 






```

# Nieuwe sectie

```python
#Part 2 time to make the plot x 1990 2005 2017 2018. y axis CO2 --> 5 countries changing over time so multiple line graph 
#years = ["1990","2005","2017","2018"]
emission.columns
fig = plt.figure()
fig, ax = plt.subplots()
emission1 = emission.nlargest(5,"2018")
data = emission1.loc[:,["country","1990","2005","2017","2018"]]
cols = data.columns[1:]

for index, row in data.iterrows():
  ax.plot(cols, row[1:], label=row[0])
plt.legend(loc=5, bbox_to_anchor=(1.4, .75))
ax.set_ylabel('CO2 emission')
ax.set_xlabel('years')
ax.set_title('CO2 of the bigger countries')
plt.show()

```

```python
#exercize 2 part 1 first relative change then part 2 x-as 1990, 2005, 2017 y-as relative amount 
emission2 = emission.loc[:,['country','1990','2005','2017','2018']]
emission2['1990p'] = emission2.loc[:,'1990'] / emission2.loc[:,'1990']*100 
emission2['change05'] = emission2.loc[:,'2005'] / emission2.loc[:,'1990']*100 
emission2['change17'] = emission2.loc[:,'2017'] / emission2.loc[:,'1990']*100 
emission2['change18'] = emission2.loc[:,'2018'] / emission2.loc[:,'1990']*100 
emission3 = emission2.sort_values(by=['change18'])
emission3 = emission3.head(3).append(emission3.tail(3))
emission3
#added 2018 sorted 2018
```

```python
#2.2 chart
dat2 = emission3.loc[:,['country','1990p','change05','change17','change18']]
col2 = dat2.columns[1:]
fig, ax = plt.subplots()
for index, row in dat2.iterrows():
  ax.plot(col2, row[1:], label=row[0])
ax.set_ylabel('percent')
ax.set_xlabel('years')
ax.set_title('change top and lowest')
plt.legend(loc=5, bbox_to_anchor=(2, .75))
plt.show()
```

```python
#with selection of other countries 
emission_mask= emission2['1990'] >= 5.000
emission4=emission2[emission_mask]
emission4 = emission4.sort_values(by=['change18'])
emission5 = emission4.head(3).append(emission4.tail(3))
dat3 = emission5.loc[:,['country','1990p','change05','change17','change18']]
col3 = dat3.columns[1:]
fig, ax = plt.subplots()
for index, row in dat3.iterrows():
  ax.plot(col3, row[1:], label=row[0])
ax.set_ylabel('percent')
ax.set_xlabel('years')
ax.set_title('change top and lowest')
plt.legend(loc=5, bbox_to_anchor=(1.5, .75))
plt.show()
```
